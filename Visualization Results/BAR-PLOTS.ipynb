```python
import matplotlib.pyplot as plt
import numpy as np

# Define metrics for the models
metrics = {
    'Random Forest': {
        'MSE': 0.0294,
        'RMSE': 0.1715,
        'MAE': 0.1320,
    },
    'Lasso Regression': {
        'MSE': 0.0313,
        'RMSE': 0.1769,
        'MAE': 0.1399,
    },
    'Linear Regression': {
        'MSE': 0.0303,
        'RMSE': 0.1741,
        'MAE': 0.1376,
    }
}

# Define the desired order of metrics
metric_order = ['RMSE', 'MAE', 'MSE']

# Prepare data for plotting
model_names = list(metrics.keys())
bar_width = 0.25
x = np.arange(len(metric_order))  # the label locations

fig, ax = plt.subplots()

# Plot bars for each model
for i, model in enumerate(model_names):
    values = [metrics[model][metric] for metric in metric_order]
    ax.bar(x + i * bar_width, values, bar_width, label=model)

# Add labels, title, and custom x-axis tick labels
ax.set_xlabel('Metrics')
ax.set_ylabel('Values')
ax.set_title('Comparison of Regression Models by Metrics')
ax.set_xticks(x + bar_width)
ax.set_xticklabels(metric_order)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define metrics for the models, including AIC and BIC only
metrics = {
    'Random Forest': {
        'AIC': -34355.586406300055,
        'BIC': -34284.80689446075,
    },
    'Lasso Regression': {
        'AIC': -29847.0015,
        'BIC': -29825.7990,
    },
    'Linear Regression': {
        'AIC': -23063.506424308413,
        'BIC': -22970.5149076114,
    }
}

# Define the desired order of metrics (AIC and BIC only)
metric_order = ['AIC', 'BIC']

# Prepare data for plotting
model_names = list(metrics.keys())
bar_width = 0.12  # Adjusted bar width for better spacing
x = np.arange(len(metric_order))  # the label locations

fig, ax = plt.subplots(figsize=(8, 5))  # Set figure size for better readability

# Plot bars for each model
for i, model in enumerate(model_names):
    values = [metrics[model][metric] for metric in metric_order]
    ax.bar(x + i * bar_width, values, bar_width, label=model)

# Add labels, title, and custom x-axis tick labels
ax.set_xlabel('Metrics', fontsize=14)
ax.set_ylabel('Values', fontsize=14)
ax.set_title('Comparison of Regression Models by AIC and BIC', fontsize=16)
ax.set_xticks(x + bar_width / 2)  # Center the ticks
ax.set_xticklabels(metric_order, fontsize=12)
ax.legend(title='Models', fontsize=12)

# Improve layout
plt.tight_layout()
plt.grid(axis='y')  # Add grid for better readability
plt.ylim(-35000, -0)  # Set y-limits to focus on the range of AIC and BIC values
plt.show()
```
